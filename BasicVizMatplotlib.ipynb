## Earnings' analysis based on college major
 *Basic visualization capabilities*

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
```

```python
# Get rid of rows with NAs

print('Before cleaning: {} records'.format(recent_grads.shape[0]))
recent_grads = recent_grads.dropna()
print('After cleaning: {} records'.format(recent_grads.shape[0]))
```

```python
# Earning vs major popularity

recent_grads.plot('Sample_size','Median',kind='scatter',title='Earnings vs Major popularity')
plt.show()
```

```python
# Earnings vs Gender Ratio in major

recent_grads.plot('ShareWomen','Median',kind='scatter',title = 'Earnings vs Women Share',xlim=(0,1))
plt.show()
```

```python
# Earnings vs Full/Part time

recent_grads.plot('Full_time','Median',kind='scatter',title = 'Earnings vs Full/Part time')
plt.show()
```

```python
# Earnings distribution

recent_grads['Median'].hist(bins=20)
plt.title('Earnings Distribution')
plt.show()
```

```python
# Men/Women share distribution

recent_grads['ShareWomen'].hist(bins=20)
plt.title('Men/Women share Distribution')
plt.xlabel('Women share')
plt.show()
```

```python
recent_grads.hist('Median',by='Major_category',figsize=(12,12),bins=10)
plt.show()
```

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(recent_grads[['Sample_size','Median']],figsize=(10,10))
plt.show()
```

```python
scatter_matrix(recent_grads[['Sample_size','Median','Unemployment_rate']],figsize=(10,10))
plt.show()
```

```python
recent_grads.iloc[:20,:].plot.bar(x='Major',y='ShareWomen',legend=None, color = 'grey')
plt.show()
```

```python
# Highest unemployment rate (top 10)

ur = recent_grads[['Major','Unemployment_rate']].sort_values(by = 'Unemployment_rate',ascending=False).head(10).reset_index(drop=True)
ur.plot.bar(x='Major',y='Unemployment_rate',legend=None,color='red')
plt.show()
```

```python
# Lowest unemployment rate (top 10)

ur = recent_grads[['Major','Unemployment_rate']].sort_values(by = 'Unemployment_rate',ascending=True).head(10).reset_index(drop=True)
ur.plot.bar(x='Major',y='Unemployment_rate',legend=None,color='green')
plt.show()
```

```python
# Men / Women by major

df = recent_grads[['Major_category','Men','Women']]
df = df.groupby('Major_category').sum().sort_values(by='Men',ascending = False)
df.plot.bar()
plt.show()
```

```python
# Median salaries vs unemployment rate and share of women in major

df = recent_grads[['Unemployment_rate','Median','ShareWomen']]

cats = pd.cut(df['Unemployment_rate'],bins=5)
cats.name = 'Unemployment_rate_bin'

cats_sw = pd.cut(df['ShareWomen'],bins=2,labels=['low','high'])
cats_sw.name = 'ShareWomen_bin'

df = pd.concat([df,cats,cats_sw],axis=1)

df.boxplot(column='Median',by=['Unemployment_rate_bin','ShareWomen_bin'],figsize=(10,6))
plt.xticks(rotation = 90)
plt.ylabel('Median salary')
plt.xlabel('Unemplyment rate bin / Women share Bin')
plt.title('Median salaries vs unemployment rate and share of women in major')
plt.show()
```
